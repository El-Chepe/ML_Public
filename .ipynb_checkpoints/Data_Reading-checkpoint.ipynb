```python
import json

data = []
with open('../businessnull.json') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
data[1:3]
```

```python
data[1]['categories']
```

```python
cities = []
for line in range(0,192608):
    cities.append(data[line]['city'])
```

```python
import pandas as pd

df = pd.DataFrame(data=cities)
```

```python
df.drop_duplicates(keep=False, inplace=True)
```
